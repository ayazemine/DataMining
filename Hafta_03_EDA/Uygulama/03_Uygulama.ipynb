# Hafta 3: Keşifsel Veri Analizi (EDA)
Veri görselleştirme ve istatistiksel analiz uygulamaları.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

## Veri Yükleme

```python
df = pd.read_csv('../../Veri_Setleri_Genel/titanic.csv')
```

## 1. Tek Değişkenli Analiz

```python
# Yaş Dağılımı (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), kde=True, bins=30)
plt.title('Yaş Dağılımı')
plt.show()
```

```python
# Hayatta Kalma Oranı (Countplot)
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df)
plt.title('Hayatta Kalma Dağılımı (0=Hayır, 1=Evet)')
plt.show()
```

## 2. İki Değişkenli Analiz

```python
# Sınıflara Göre Yaş Dağılımı (Boxplot)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Yolcu Sınıfına Göre Yaş Dağılımı')
plt.show()
```

## 3. Korelasyon Analizi

```python
# Sadece sayısal sütunları seçelim
numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelasyon Matrisi')
plt.show()
```
