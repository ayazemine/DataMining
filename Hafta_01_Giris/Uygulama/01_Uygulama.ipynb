# Hafta 1: Python ve Veri Bilimi Temelleri
Bu notebook, Python veri yapıları, Numpy ve Pandas kütüphanelerine giriş niteliğindedir.

```python
import numpy as np
import pandas as pd
```

## 1. Python Veri Yapıları

```python
# Listeler
my_list = [1, 2, 3, 4, 5]
print("Liste:", my_list)

# Sözlükler
my_dict = {"isim": "Ahmet", "yas": 25, "meslek": "Muhendis"}
print("Sözlük:", my_dict)
```

## 2. Kontrol Yapıları (If-Else, Döngüler)

```python
# If-Else Yapısı
puan = 75
if puan >= 50:
    print(f"Puanınız: {puan} - Geçti")
else:
    print(f"Puanınız: {puan} - Kaldı")

# For Döngüsü
print("\nİsimler Listesi:")
for i, isim in enumerate(my_list):
    print(f"{i}: {isim}")
```

## 3. Fonksiyonlar

```python
def kare_al(x):
    return x ** 2

print("4'ün karesi:", kare_al(4))

lamb = lambda x: x * 10
print("Lambda ile 5*10:", lamb(5))
```

## 4. Numpy Dizileri

```python
arr = np.array([10, 20, 30, 40])
print("Numpy Array:", arr)
print("Ortalama:", np.mean(arr))
print("Standart Sapma:", np.std(arr))
```

## 5. Pandas DataFrames

```python
data = {
    'Isim': ['Ali', 'Veli', 'Ayse', 'Fatma'],
    'Not': [85, 90, 78, 92],
    'Yas': [20, 21, 20, 22]
}
df = pd.DataFrame(data)
df
```

```python
# Temel Bilgiler
print(df.info())
print(df.describe())
```

## 5.1. Pandas ile Veri Manipülasyonu

```python
# Filtreleme: Notu 80'den büyük olanlar
yuksek_not = df[df['Not'] > 80]
print("Yüksek Notlu Öğrenciler:\n", yuksek_not)

# Yeni Sütun Ekleme (Durum)
df['Durum'] = ["Geçti" if x >= 80 else "Kaldı" for x in df['Not']]
print("\nYeni Sütunlu Tablo:\n", df)
```

## 6. Basit Görselleştirme (Matplotlib Entegrasyonu)

```python
import matplotlib.pyplot as plt

# Basit bir bar grafiği
df.plot(x='Isim', y='Not', kind='bar', color='salmon', legend=False)
plt.title('Öğrenci Notları')
plt.ylabel('Not')
plt.show()
```
