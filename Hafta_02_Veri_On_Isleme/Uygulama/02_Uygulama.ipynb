# Hafta 2: Veri Ön İşleme
Bu notebookta kayıp veri analizi, aykırı değer tespiti ve normalizasyon işlemleri yapılacaktır.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

## Veri Setini Yükleme

```python
df = pd.read_csv('../../Veri_Setleri_Genel/titanic.csv')
df.head()
```

## 1. Kayıp Veri Analizi

```python
print(df.isnull().sum())
```

```python
# 'Age' sütunundaki eksiklikleri medyan ile dolduralım
df['Age'].fillna(df['Age'].median(), inplace=True)

# 'Cabin' sütununu çok fazla eksik olduğu için silelim
if 'Cabin' in df.columns:
    df.drop('Cabin', axis=1, inplace=True)

print("Doldurma ve silme sonrası eksiklikler:")
print(df.isnull().sum())
```

## 2. Kategorik Dönüşüm (Encoding)

```python
# Cinsiyet (Sex) sütununu map edelim: male=0, female=1
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df.head()
```

## 3. Ölçeklendirme (Scaling)

```python
# Age ve Fare sütunlarını normalize edelim (Min-Max)
scaler = MinMaxScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
df.head()
```
